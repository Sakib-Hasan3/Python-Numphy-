# Uniform Distribution

```python
# low - lower bound - default 0 .0.

# high - upper bound - default 1.0.

# size - The shape of the returned array.
```

```python
from numpy import random
x=random.uniform(size=(2,3,)) #By default, it generates values in the range [0,1] from a uniform distribution.
print(x)
```

```python
from numpy import random
x=random.uniform(l=10,high=20,size=(2,3,)) #By default, it generates values in the range [0,1] from a uniform distribution.
print(x)
```

# visulaization of uniform distribution

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.uniform(size=1000), hist=False)

plt.show()
```
